## Barplot

```python
diamonds = sns.load_dataset('diamonds')
dataFrameDia = diamonds.copy()
dataFrameDia.head()
```

```python
dataFrameDia.shape
```

```python
dataFrameDia.dtypes
```

```python
dataFrameDia.describe().T
```

```python
from pandas.api.types import CategoricalDtype
dataFrameDia.cut = dataFrameDia.cut.astype(CategoricalDtype(ordered = True))
dataFrameDia.color = dataFrameDia.color.astype(CategoricalDtype(ordered = True))
dataFrameDia.clarity= dataFrameDia.clarity.astype(CategoricalDtype(ordered = True))
dataFrameDia.dtypes
```

```python
dataFrameDia.cut.head()
```

```python
cutCat = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
dataFrameDia.cut = dataFrameDia.cut.astype(CategoricalDtype(cutCat, ordered = True))
dataFrameDia.cut.head()
```

```python
dataFrameDia['cut'].value_counts().plot.barh().set_title('Barplot')
```

```python
dataFrameDia['cut'].value_counts().plot.barh().set_title('Barplot');
```

```python
(dataFrameDia['cut']
 .value_counts()
 .plot.barh()
 .set_title('Barplot'));
```

```python
(dataFrameDia['cut']
 .value_counts()
 .plot.barh()
 .set_title('Barplot'));
```

```python
sns.catplot(x = 'cut', y = 'carat', data = dataFrameDia)
```

```python
sns.catplot(x = 'cut', y = 'price', data = dataFrameDia)
```

```python
sns.catplot(x = 'cut', y = 'price', hue = "color", data = dataFrameDia)
```
