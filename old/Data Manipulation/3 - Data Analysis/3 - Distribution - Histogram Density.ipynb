## Distribution - Histogram Density

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
(diamonds['price']
 .plot
 .hist(bins=100)
)
```

```python
sns.distplot(diamonds.price, kde = False)
```

```python
sns.distplot(diamonds.price)
```

```python
sns.distplot(diamonds.price, hist = False)
```

```python
sns.kdeplot(diamonds.price, shade = True)
```

```python
(sns
.FacetGrid(diamonds, hue ='cut', height=5, xlim=(0, 10000))
.map(sns.kdeplot, 'price', shade=True)
.add_legend()
)
```

```python
sns.catplot(x = 'cut', y = 'price', hue = 'color', kind = 'point', data = diamonds);
```
